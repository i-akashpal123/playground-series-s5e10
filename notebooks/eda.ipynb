```python
##. Target Variable Analysis


import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(train_data['accident_risk'],kde=True,color='pink')
plt.title("Distribution of target : Accident Risk")
plt.show()
```

```python
train_data.dtypes
```

```python
numeric_col_train_data = train_data.select_dtypes(include=['int64','float64'])
```

```python
corr=numeric_col_train_data.corr()['accident_risk'].sort_values(ascending=False)
```

```python
corr
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=corr.head(10).index, y=corr.head(10).values)
plt.title("Top 10 Numeric Features Correlated with Accident Risk")
plt.xticks(rotation=45)
plt.show()
```

```python
#Visualize Feature vs Target Relationships

top_features = ['curvature', 'speed_limit', 'num_reported_accidents']

for col in top_features:
    sns.scatterplot(x=train_data[col], y=train_data['accident_risk'], alpha=0.3)
    plt.title(f'{col} vs Accident Risk')
    plt.show()
```

```python

import numpy as np
cat_features = ['road_type', 'lighting', 'weather', 'time_of_day']

for col in cat_features:
    plt.figure(figsize=(6,4))
    sns.barplot(x=col, y='accident_risk', data=train_data, estimator=np.mean, ci=None, palette='Set2')
    plt.title(f'Average Accident Risk by {col}')
    plt.xticks(rotation=45)
    plt.show()
```

### EDA Summary

**Numeric Features**
- `curvature`: Strong positive correlation with accident risk.
- `speed_limit`: Moderate positive relationship; risk slightly increases at higher speed limits.
- `num_reported_accidents`: Weak-to-moderate positive correlation; areas with past accidents remain risky.

**Categorical Features**
- `lighting`: Strongest effect among categorical variables; accidents rise significantly at night.
- `weather`: Foggy and rainy conditions contribute to higher accident risk.
- `road_type`: Slightly higher risk on urban roads.
- `time_of_day`: Minimal variation; less predictive power.

**Modeling Implications**
- Perform one-hot encoding for categorical columns (`road_type`, `lighting`, `weather`, `time_of_day`).
- Scale numeric columns if using distance-based models (optional for tree models).
- Expect `curvature`, `speed_limit`, and `lighting` to be top predictive drivers.
